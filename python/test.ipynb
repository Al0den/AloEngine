```python
import json
import numpy as np
```

```python
def find_max_abs_cp(jsonl_path):
    max_abs = 0.0
    with open(jsonl_path, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            obj = json.loads(line)
            cp = abs(float(obj["eval"]))
            if cp > max_abs:
                max_abs = cp
    return max_abs
    
path = "../data/nnue_dataset.jsonl"
m = find_max_abs_cp(path)
```

```python
print(m)
```

```python
vals=[]
with open("../data/nnue_dataset.jsonl") as f:
    for line in f:
        vals.append(json.loads(line)["eval"])
vals=np.array(vals)

print("min =", vals.min())
print("max =", vals.max())
print("mean =", vals.mean())
print("abs-mean =", np.abs(vals).mean())
print("95th percentile =", np.percentile(np.abs(vals),95))
```
